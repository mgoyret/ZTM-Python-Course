tomar solo onmbres, salarios y valores de los jugadores

```python
data_frame.loc[1:10, ['Name', 'Value', 'Wage']]
```

```python
new_df = pd.DataFrame(data_frame, columns=['Name', 'Value', 'Wage'])

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = new_df['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
value = new_df['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)
new_df['Wage']=wage
new_df['Value']=value
```
